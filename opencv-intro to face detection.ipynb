```python
#face recognition
test_image = cv2.imread('baby1.jpg')
test_image_gryscale = cv2.cvtColor(test_image,cv2.COLOR_BGR2GRAY)
plt.imshow(test_image_gryscale, cmap='gray')
```

```python
def convertToRGB(image):
    return cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
haar_cascade_face = cv2.CascadeClassifier('haarcascades/haarcascade_frontalface_default.xml')
```

```python
faces_rects = haar_cascade_face.detectMultiScale(test_image_gryscale, scaleFactor = 1.2, minNeighbors = 5);
print('Faces Found: ', len(faces_rects))
```

```python
for (x,y,w,h) in faces_rects:
    cv2.rectangle(test_image,(x,y),(x+w, y+h), (0,255,0), 2)
```

```python
plt.imshow(convertToRGB(test_image))
```

```python
#face recognition
test_image = cv2.imread('group1.jpg')
test_image_gryscale = cv2.cvtColor(test_image,cv2.COLOR_BGR2GRAY)
plt.imshow(test_image_gryscale, cmap='gray')
haar_cascade_face = cv2.CascadeClassifier('haarcascades/haarcascade_frontalface_default.xml')
faces_rects = haar_cascade_face.detectMultiScale(test_image_gryscale, scaleFactor = 1.2, minNeighbors = 5);
print('Faces Found: ', len(faces_rects))
```

```python
for (x,y,w,h) in faces_rects:
    cv2.rectangle(test_image,(x,y),(x+w, y+h), (0,255,0), 2)
plt.imshow(convertToRGB(test_image))
```
